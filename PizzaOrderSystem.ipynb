```python
#Sisteme giriş ve çıkış yazıları
def welcome_text():
    print("Pizza Sipariş Sistemi'ne Hoşgeldiniz")
def goodbye_text():
    print("Pizza sipariş sistemini kullandığınız için teşekkür ederim.")
```

```python
# Pizza menüsünü yazdırma
pizza_menu = open("Menu.txt", "r")
def display_menu():
    print(pizza_menu.read())
```

```python
#Pizza sınıfı
class Pizza():
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
```

```python
# Pizza Alt Sınıfları
class ClasicPizza(Pizza):
    def __init__(self):
        super().__init__(description= "Klasik Pizza",cost= 99.95)

class Margarita(Pizza):
    def __init__(self):
        super().__init__(description="Margarita Pizza",cost= 96.95)

class TurkPizza(Pizza):
    def __init__(self):
        super().__init__(description="Türk Pizza",cost= 89.95)

class SadePizza(Pizza):
    def __init__(self):
        super().__init__(description="Sade Pizza",cost= 75.00)        
```

```python
#Boyut Üst Sınıfı
class SizeDecorator(Pizza):
    def __init__(self, pizza):
       self.pizza = pizza
    
    def get_description(self):
        return self.pizza.get_description()
    
    def get_cost(self):
        return self.pizza.get_cost()
```

```python
#Boyut Alt Sınıfları
class Small(SizeDecorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 0.00
        self.description = "Küçük Boy "

    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Medium(SizeDecorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 9.90
        self.description = "Orta Boy "

    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Large(SizeDecorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 16.90
        self.description = "Büyük Boy "

    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
```

```python
#Sos(Ekstra Malzeme) Üst Sınıfı
class ToppingDecarotor(Pizza):
    def __init__(self, pizza):
        self.pizza = pizza

    def get_description(self):
        return self.pizza.get_description()
    
    def get_cost(self):
        return self.pizza.get_cost()
        
```

```python
#Sos(Ekstra Malzeme) Alt Sınıfları
class Zeytin(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 9.90
        self.description = " Ekstra Zeytin"
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Mantar(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 11.90
        self.description = " Ekstra Mantar"
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Peynir(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 14.90
        self.description = " Ekstra Keçi Peyniri"
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Et(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 19.90
        self.description = " Ekstra Et"
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Sogan(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 6.90
        self.description = " Ekstra Soğan"
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost

class Misir(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 8.90
        self.description = " Ekstra Mısır"
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class NoExtra(ToppingDecarotor):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.cost = 0.00
        self.description = " "
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
```

```python
#Siparişin veritabanına yazdırılması
import csv
import os
from datetime import datetime

def write_order_to_csv(name, id_no, order_description, total_cost, card_no, card_cvv):
    #Sipariş zamanının alınması
    now = datetime.now()
    order_time = now.strftime("%d/%m/%Y, %H:%M:%S")

    #Yeni siparişin oluşturulması
    new_order = [name, id_no, order_description, "%.2f" % total_cost, order_time, card_no, card_cvv]

    file_exists = os.path.isfile("Orders_Database.csv")

    #Orders_Database.csv dosyasına sipariş satırının eklenmesi ve yoksa başlığın eklenmesi
    with open("Orders_Database.csv", mode="a", newline="") as file:
        writer = csv.writer(file)
        if file_exists:
            writer.writerow(new_order)
        else:
            header = (["Ad-Soyad", "TC Kimlik No", "Sipariş Açıklaması", "Sipariş Tutarı", "Sipariş Zamanı", "Kart No", "Kart CVV"])
            writer.writerow(header)
            writer.writerow(new_order)
```

```python
#Sipariş sisteminin çalışmasını sağlayacak main() fonksiyonu
def main():

    welcome_text()
    display_menu()

    # Kullanıcıdan pizza seçimini alma
    chosen_pizza = input("Lütfen pizza seçiniz(1-4):")
    while chosen_pizza not in ["1", "2", "3", "4"]:
        chosen_pizza = input("Lütfen geçerli bir pizza numarası seçiniz(1-4):")
    chosen_pizza = int(chosen_pizza)

    #Seçilen pizzayı tanımlama
    if chosen_pizza == 1:
        pizza = ClasicPizza()
    elif chosen_pizza == 2:
        pizza = Margarita()
    elif chosen_pizza == 3:
        pizza = TurkPizza()
    elif chosen_pizza == 4:
        pizza = SadePizza()

    #Kullanıcıdan pizza büyüklük seçimini alma
    chosen_size = input("Lütfen pizzanızın boyutunu seçiniz(5-7):")
    while chosen_size not in ["5", "6", "7"]:
        chosen_size = input("Lütfen geçerli bir boyut numarası seçiniz(1-4):")
    chosen_size = int(chosen_size)

    #Seçilen büyüklüğü tanımlama
    if chosen_size == 5:
        size = Small(pizza)
    elif chosen_size == 6:
        size = Medium(pizza)
    elif chosen_size == 7:
        size = Large(pizza)
    
    #Kullanıcıdan ekstra malzeme(sos) seçimini kaç ekstra istediğine göre alma
    numberOfTopping = input("Kaç adet ekstra malzeme istiyorsunuz? (0-3)")
    while numberOfTopping not in ["1", "2", "3", "0"]:
        numberOfTopping = input("Lütfen 0 ile 3 arasında bir sayı giriniz: ")
    numberOfTopping = int(numberOfTopping)
    topping = [NoExtra(pizza), NoExtra(pizza), NoExtra(pizza)]
    
    #Döngü ile kullanıcın istediği kadar malzeme seçim kodu çalışıyor
    while numberOfTopping > 0:
        
        chosen_topping = input("Lütfen ekstra malzeme seçiniz (11-17)")
        while chosen_topping not in ["11", "12", "13", "14", "15", "16", "17"]:
            chosen_topping = input("Lütfen geçerli bir malzeme numarası seçin (11-17): ")
        chosen_topping = int(chosen_topping)
        

        if chosen_topping == 11:
            topping[numberOfTopping-1] = Zeytin(pizza)
        elif chosen_topping == 12:
            topping[numberOfTopping-1] = Mantar(pizza)
        elif chosen_topping == 13:
            topping[numberOfTopping-1] = Peynir(pizza)
        elif chosen_topping == 14:
            topping[numberOfTopping-1] = Et(pizza)
        elif chosen_topping == 15:
            topping[numberOfTopping-1] = Sogan(pizza)
        elif chosen_topping == 16:
            topping[numberOfTopping-1] = Misir(pizza)
        elif chosen_topping == 17:
            topping[numberOfTopping-1] = NoExtra(pizza)
        
        numberOfTopping -=1
        
    # Kullanıcı bilgilerinin alınması
    name= input("Adınız-Soyadınız: ")
    id_no= input("TC Kimlik Numaranız: ")
    card_no= input("Kredi/Banka Kartı Numaranız: ")
    card_cvv= input("Kartınızın CVV numarası: ")

    # Sipariş açıklamasının oluşturulması
    order_description = size.get_description() + pizza.get_description() + topping[0].get_description() + topping[1].get_description() + topping[2].get_description()
    
    #Sipariş fiyatının hesaplanması
    total_cost = pizza.get_cost() + size.get_cost() + topping[0].get_cost() + topping[1].get_cost() + topping[2].get_cost() 
    print("\n**** SİPARİŞ ÖZETİ ****")
    print(order_description)
    print("Toplam Fiyat \t" "%.2f" % total_cost)
    
    # Siparişin veritabanına kaydedilmesi
    write_order_to_csv(name, id_no, order_description, total_cost, card_no, card_cvv)
    goodbye_text()

```

```python
# Main fonksiyon
main()
```
